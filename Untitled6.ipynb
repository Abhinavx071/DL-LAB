```python
import os
import re
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from collections import Counter
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
```

```python
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
OUTPUT_DIR = os.path.join(os.getcwd(), "output")
os.makedirs(OUTPUT_DIR, exist_ok=True)

MAX_VOCAB = 20000
MAX_LEN = 300
BATCH_SIZE = 64
```

```python
def tokenize(text):
    text = text.lower()
    text = re.sub(r"<br\s*/?>", " ", text)
    return re.findall(r"\b\w+\b", text)


def load_imdb():
    from datasets import load_dataset
    ds = load_dataset("imdb")

    train_texts  = ds["train"]["text"]
    train_labels = ds["train"]["label"]
    test_texts   = ds["test"]["text"]
    test_labels  = ds["test"]["label"]

    return train_texts, train_labels, test_texts, test_labels
```

```python
def build_vocab(texts, max_vocab):
    counts = Counter()
    for t in texts:
        counts.update(tokenize(t))

    word2idx = {"<PAD>": 0, "<UNK>": 1}
    for word, _ in counts.most_common(max_vocab - 2):
        word2idx[word] = len(word2idx)
    return word2idx


def encode(text, word2idx, max_len):
    ids = [word2idx.get(tok, 1) for tok in tokenize(text)[:max_len]]
    ids += [0] * (max_len - len(ids))
    return ids
```

```python
class IMDBDataset(Dataset):
    def __init__(self, texts, labels, word2idx, max_len):
        self.texts   = texts
        self.labels  = labels
        self.word2idx = word2idx
        self.max_len  = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        x = torch.tensor(encode(self.texts[idx], self.word2idx, self.max_len), dtype=torch.long)
        y = torch.tensor(self.labels[idx], dtype=torch.long)
        return x, y
```

```python
class SentimentRNN(nn.Module):
    def __init__(self, vocab_size, embed_dim=128, hidden_dim=128, num_layers=2, num_classes=2, dropout=0.3):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.rnn = nn.RNN(embed_dim, hidden_dim, num_layers=num_layers,
                          batch_first=True, dropout=dropout)
        self.drop = nn.Dropout(dropout)
        self.fc   = nn.Linear(hidden_dim, num_classes)

    def forward(self, x):
        emb = self.embedding(x)
        _, hidden = self.rnn(emb)
        out = self.fc(self.drop(hidden[-1]))
        return out
```

```python
def train_epoch(model, loader, criterion, optimizer):
    model.train()
    total_loss, correct, n = 0.0, 0, 0

    for x, y in loader:
        x, y = x.to(DEVICE), y.to(DEVICE)
        optimizer.zero_grad()
        logits = model(x)
        loss = criterion(logits, y)
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 5.0)
        optimizer.step()

        total_loss += loss.item() * len(y)
        correct += (logits.argmax(1) == y).sum().item()
        n += len(y)

    return total_loss / n, correct / n
```

```python
@torch.no_grad()
def evaluate(model, loader, criterion):
    model.eval()
    total_loss, correct, n = 0.0, 0, 0

    for x, y in loader:
        x, y = x.to(DEVICE), y.to(DEVICE)
        logits = model(x)
        loss = criterion(logits, y)

        total_loss += loss.item() * len(y)
        correct += (logits.argmax(1) == y).sum().item()
        n += len(y)

    return total_loss / n, correct / n
```

```python
def main():
    print(f"using device: {DEVICE}")
    print("loading data...")
    train_texts, train_labels, test_texts, test_labels = load_imdb()

    print("building vocab...")
    word2idx = build_vocab(train_texts, MAX_VOCAB)
    print(f"vocab size: {len(word2idx)}")

    train_ds = IMDBDataset(train_texts, train_labels, word2idx, MAX_LEN)
    test_ds  = IMDBDataset(test_texts,  test_labels,  word2idx, MAX_LEN)
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE)

    model     = SentimentRNN(len(word2idx)).to(DEVICE)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

    EPOCHS = 10
    history = {"train_loss": [], "train_acc": [], "test_loss": [], "test_acc": []}

    for epoch in range(1, EPOCHS + 1):
        tr_loss, tr_acc = train_epoch(model, train_loader, criterion, optimizer)
        te_loss, te_acc = evaluate(model, test_loader, criterion)

        history["train_loss"].append(tr_loss)
        history["train_acc"].append(tr_acc)
        history["test_loss"].append(te_loss)
        history["test_acc"].append(te_acc)

        print(f"[{epoch}/{EPOCHS}]  loss {tr_loss:.4f}/{te_loss:.4f}  acc {tr_acc:.4f}/{te_acc:.4f}")

    # plots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    ax1.plot(history["train_loss"], label="train")
    ax1.plot(history["test_loss"],  label="test")
    ax1.set_title("loss")
    ax1.set_xlabel("epoch")
    ax1.legend()

    ax2.plot(history["train_acc"], label="train")
    ax2.plot(history["test_acc"],  label="test")
    ax2.set_title("accuracy")
    ax2.set_xlabel("epoch")
    ax2.legend()

    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "training_curves.png"), dpi=150)
    plt.show()  # renders inline in notebook

    label_map = {0: "negative", 1: "positive"}
    model.eval()

    with open(os.path.join(OUTPUT_DIR, "sample_predictions.txt"), "w", encoding="utf-8") as f:
        f.write(f"final test acc: {history['test_acc'][-1]:.4f}\n\n")

        for i, (text, true) in enumerate(zip(test_texts[:10], test_labels[:10])):
            ids = torch.tensor([encode(text, word2idx, MAX_LEN)], dtype=torch.long).to(DEVICE)
            with torch.no_grad():
                pred = model(ids).argmax(1).item()

            f.write(f"--- sample {i+1} ---\n")
            f.write(f"{text[:200]}...\n")
            f.write(f"true: {label_map[true]}  |  pred: {label_map[pred]}\n\n")

    torch.save(model.state_dict(), os.path.join(OUTPUT_DIR, "rnn_sentiment_model.pth"))
    print(f"outputs saved to {OUTPUT_DIR}")
```

```python
main()
```
