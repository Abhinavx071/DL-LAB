WHY TENSORS


```python
import time
N = 100000
lista = list(range(N))
listb = list(range(N))

na = np.array(lista)
nb = np.array(listb)
```

```python
start = time.perf_counter()
dlist = 0
for i in range(N):
  dlist+=lista[i]*listb[i]
end = time.perf_counter()

print(f"Time for list {end-start}")
```

```python
start = time.perf_counter()
dot_numpy = np.dot(na , nb)
end = time.perf_counter()

print(f"Time for numpy {end-start}")
```

SHAPE MANIPULATION

```python
x = torch.arange(6)
print(x)

y = x.view(2 , 3)
print(y)

z = x.reshape(2 , 3)
print(z)
```

```python
#UNSQUEEZE -> ADDS DIMENSION
x = torch.tensor([1, 2, 3])
print(x.shape)

y = x.unsqueeze(0)
print(y.shape)

z = x.unsqueeze(1)
print(z.shape)
```

```python
#SQUEEZE -> REMOVES DIMENSION
a = torch.randn(1, 3, 1, 5)
print(a.shape)

b = a.squeeze()
print(b.shape)
```

DIFFERENCE BETWEEN RESHAPE OF PYTORCH AND NUMPY


```python
#NUMPY
x = np.arange(6).reshape(2, 3)
y = x.T
z = y.reshape(6)

print(z)
```

```python
#PYTORCH
x = torch.arange(6).reshape(2, 3)
y = x.t()

z = y.view(6)
#gives error as memory is non contiguous
```

BROADCASTING


```python
#SCALAR

x = torch.tensor([1, 2, 3])
y = x + 10
print(y)
```

```python
#VECTOR
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

y = torch.tensor([10, 20, 30])

z = x + y
print(z)
```

In-place vs Out-of-place Operations

```python
#Out of space -> creates new tensors
#In place -> modifies at same memory location
x = torch.tensor([1., 2., 3.])

y = x + 1
print(x)
print(y)
```

```python
x = torch.tensor([1., 2., 3.])
y = x

x.add_(1)
print(y)
```
